```python
#load a dataset
import pandas as pd
df = pd.read_csv('amazon_top1000_customers_2019_2024.csv')
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
#missing values
df.isnull().sum()
```

```python
#year-wise sales analysis
df.groupby("year")["sales_amount"].sum().plot(kind='line',title='Year-wise Sales')
```

```python
#category-wise sales analysis
df.groupby("category")["sales_amount"].sum().plot(kind='bar',title='Category')
```

```python
#payment mode analysis
df['payment_mode'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```

```python
#city wise revenue analysis
df.groupby("city")["sales_amount"].sum().sort_values(ascending=False).head(10)
```

```python
#outliers detection box plot
import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(x=df['sales_amount'])
plt.title('Sales Amount outliers')
plt.show()
```
